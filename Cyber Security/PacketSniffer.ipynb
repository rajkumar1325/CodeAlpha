```python
# Scapy:: packet manipulation tool and library, used for network packet crafting, sniffing, and analysis.
from scapy.all import *
```

# We are going to check for basic Protocols :
1. TCP {Transport Layer Protocols} :: reliable communication between devices on a network.

2. IP {Internet protocol} :: identify devices on the network and to route packets to the correct destination.

3. UDP {User Datagram Protocol} :: unreliable communication between devices on a network.

4. ARP {Address Resolution Protocol} :: Resolves IP addresses to MAC addresses within a local network.

5. Ether {Ethernet} :: defining how data is transmitted on a wired network.







```python

# function to process and analyze the captured packets
def processPacket(packet):

    # Checks if the packet contains an IP layer.
    if packet.haslayer(IP):
        ip_src = packet[IP].src  # Source IP address            packet[IP]: Accesses the IP layer in the packet.ll;;lkk
        ip_dst = packet[IP].dst  # Destination IP address
        print(f"IP Packet is from: {ip_src} -> {ip_dst}")     #source_IP_Address--> Destination_IP_Address

    # Check if the packet has a TCP layer
    if packet.haslayer(TCP):
        tcp_src_port = packet[TCP].sport  # Source port
        tcp_dst_port = packet[TCP].dport  # Destination port
        print(f"TCP Packet is from: {tcp_src_port} -> {tcp_dst_port}")

    # Check if the packet has a UDP layer
    if packet.haslayer(UDP):
        udp_src_port = packet[UDP].sport  # Source port
        udp_dst_port = packet[UDP].dport  # Destination port
        print(f"UDP Packet: {udp_src_port} -> {udp_dst_port}")

    # Check if the packet has an ARP layer
    if packet.haslayer(ARP):
        arp_op = packet[ARP].op  # ARP operation (Request/Reply)
        print(f"ARP Packet: {arp_op} - {packet[ARP].psrc} -> {packet[ARP].pdst}")

    # Check for Ethernet frames
    if packet.haslayer(Ether):
        eth_src = packet[Ether].src  # Ethernet source MAC address
        eth_dst = packet[Ether].dst  # Ethernet destination MAC address
        print(f"Ethernet Frame: {eth_src} -> {eth_dst}")

    print('\n')
```

```python
print("Network sniffing started... Capturing First 50 packets")
sniff(prn=processPacket, store=0, count=50)
# sniff()            fn from scapy to capture packets.
# prn=processPacket  This specifies the function (process_packet) that will be called for each packet captured by sniff().
# store=0            This tells Scapy not to store the captured packets in memory.{default, sniff() stores all packets it captures in a list.}
# count              specifies how many packets to capture before stopping. {count=0 --> infinite}
```
